```python
# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
# url_for() function.

#ANSWER: In Flask, the url_for() function is used for URL building. The url_for() function takes the name of the view function as its first argument,
# and any additional arguments or keyword arguments needed to generate the URL.
#It returns a URL for the specified view function based on the URL routing rules defined in the Flask application.


from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return "<h1>Hello this is the index page<h1/>"

@app.route('/profile/<username>')
def profile(username):
    return f"Hello, {username}!!"

if __name__=="__main__":
    with app.test_request_context():
        print(url_for('index'))
        print(url_for('profile', username = "Amandeep"))
    app.run(host='0.0.0.0')    
```
