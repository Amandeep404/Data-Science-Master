### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Answer 1-  heatmap, lineplot, barplot, histplot, displot, boxplot

```python
import seaborn as sns
```

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

```python
df = sns.load_dataset('fmri')
```

```python
df.head()
```

```python
sns.lineplot(data = df, x=df['timepoint'], y = df['signal'], hue = df['region'], style = df['event'])
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.



```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,8), num=2)
plt.subplot(121)
sns.boxplot(x = df.pclass, y = df.age)
plt.subplot(122)
sns.boxplot(x = df.pclass, y = df.age)
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue  parameter for the 'cut' column of the diamonds dataset.


```python
df = sns.load_dataset('diamonds')
```

```python
df.head()
```

```python
sns.histplot(x = df.price, hue = df.cut)
```

### Q6. Use the "flights" dataset from seaborn to plot a heatmap.

```python
df = sns.load_dataset('flights')
```

```python
df.head()
```

```python
sns.heatmap(df, annot=True, fmt='d',cmap='YlGnBu')
```

This will show error so we need to pivot the table to create a matrix for the heatmap
