```python
import numpy as np
import matplotlib.pyplot as plt
```

Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a popular data visualization library in Python that provides a wide range of tools for creating high-quality static, animated, and interactive visualizations in Python.

Matplotlib is used for a variety of applications, including exploratory data analysis, statistical analysis, scientific visualization, and creating publication-quality graphics. It allows you to create a wide range of plots such as :
* line plots
* scatter plots
* bar plot
* histogram plots
* pie charts

```python
'''Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.'''

'''Answer - A scatter plot is a type of plot in data visualization that displays the relationship between two 
continuous variables. The plot consists of a collection of points, each representing the value of the two variables. 
The horizontal axis represents one variable, while the vertical axis represents the other variable.'''

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("This is x-axis")
plt.ylabel("This is y-axis")
```

```python
'''Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.'''
'''Answer - The subplot() function in Matplotlib is used to create multiple plots in a single figure.
It allows you to display several plots together, which is useful when you want to compare different plots or display 
related data in a single figure.'''

fig, axes = plt.subplots(nrows=2,ncols=2)
ax1, ax2, ax3, ax4 = axes.flatten()

x = np.array([0, 1, 2, 3, 4, 5])  
y = np.array([0, 100, 200, 300, 400, 500])
ax1.plot(x,y)
ax2.plot(np.array([0, 1, 2, 3, 4, 5]),np.array([50, 20, 40, 20, 60, 70]))
ax3.plot(np.array([0, 1, 2, 3, 4, 5]),np.array([10, 20, 30, 40, 50, 60]))
ax4.plot(np.array([0, 1, 2, 3, 4, 5]), np.array([200, 350, 250, 550, 450, 150]))
plt.show()

```

```python
'''Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.'''

'''Answer-  Bar plot, also known as a bar chart or bar graph, is a type of chart that displays rectangular bars 
with lengths proportional to the values they represent. The bars can be either horizontal or vertical.
Bar plots are used to compare the values of different categories or groups.
They are particularly useful for showing how a set of values changes over time, or for comparing the size of different values or groups. '''

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.title("Company Profits Bar Plot")
plt.xlabel("Companies")
plt.ylabel("Profit")

plt.figure()
plt.barh(company, profit)
plt.title("Company Profits Horizontal plot")
plt.xlabel("Profit")
plt.ylabel("Companies")

```

```python
'''Q5: What is a box plot? Why is it used? Using the following data plot a box plot.'''


box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1, box2])
plt.show()
```
